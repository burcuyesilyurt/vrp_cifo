```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import chardet
import seaborn as sns
```

```python
#read excel file
df = pd.read_excel(os.path.join('CIFO2024', 'evolution_data.xlsx'))
df.head()
```

```python
grouped = df.groupby(['Selection', 'Crossover', 'Mutation', 'Generation'])
mean_fitness = grouped['Fitness'].mean()

fig, ax = plt.subplots(figsize=(18, 10))  

mean_fitness.unstack(level=[0,1,2]).plot(ax=ax, linewidth=2)  

plt.xlabel('Generation')
plt.ylabel('Mean Fitness')
plt.title('Mean Fitness Values for Each Generation Grouped by Selection, Crossover and Mutation')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
grouped = df.groupby(['Selection', 'Crossover', 'Mutation', 'Generation'])
mean_fitness = grouped['Fitness'].median()

fig, ax = plt.subplots(figsize=(18, 10))  

mean_fitness.unstack(level=[0,1,2]).plot(ax=ax, linewidth=2)  

plt.xlabel('Generation')
plt.ylabel('Median Fitness')
plt.title('Median Fitness Values for Each Generation Grouped by Selection, Crossover and Mutation')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
df['Group'] = df['Selection'] + '-' + df['Crossover'] + '-' + df['Mutation']

last_gen = df[df['Generation'] == 50]
sns.set(style="whitegrid")

fig, ax = plt.subplots(figsize=(17, 8))

sns.boxplot(x='Group', y='Fitness', data=last_gen, palette="Set3", ax=ax)
plt.title('Boxplot of Fitness Values for Each Group', fontsize=16)
plt.suptitle('')
plt.xticks(rotation=45, ha='right')
plt.show()
```
