### Parameters

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import chardet
import seaborn as sns
import os
```

### Read files

```python
#print current path
print(os.getcwd())
```

```python
#read excel file
df = pd.read_excel(os.path.join('evolution_data lc101.xlsx')) #Instance evolution_data lc101
df3 = pd.read_excel(os.path.join( 'evolution_data lc101 with elitism.xlsx')) #Instance evolution_data lc101 with elitism
df4 = pd.read_excel(os.path.join('evolution_data c202C16.xlsx')) #Instance evolution_data c202C16 
```

```python
df.head()
```

#### Plot the mean fitness values for  lc101 instance

```python
#Is like not igual 
df_random= df[df['Crossover'].str.contains('random', case=False, na=False)]
#Not random
df_not_random= df[~df['Crossover'].str.contains('random', case=False, na=False)]
df_random.head()
```

```python
grouped = df_random.groupby(['Selection', 'Crossover', 'Mutation', 'Generation'])
mean_fitness = grouped['Fitness'].mean()

fig, ax = plt.subplots(figsize=(18, 10))  

mean_fitness.unstack(level=[0,1,2]).plot(ax=ax, linewidth=2)  

plt.xlabel('Generation')
plt.ylabel('Mean Fitness')
plt.title('Mean Fitness Values for Each Generation Grouped by Selection, Crossover and Mutation')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
grouped = df_not_random.groupby(['Selection', 'Crossover', 'Mutation', 'Generation'])
mean_fitness = grouped['Fitness'].mean()

fig, ax = plt.subplots(figsize=(18, 10))  

mean_fitness.unstack(level=[0,1,2]).plot(ax=ax, linewidth=2)  

plt.xlabel('Generation')
plt.ylabel('Mean Fitness')
plt.title('Mean Fitness Values for Each Generation Grouped by Selection, Crossover and Mutation')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

#### Boxplot of fitness value for  lc101 instance

```python
df_random['Group'] = df_random['Selection'] + '-' + df_random['Crossover'] + '-' + df_random['Mutation']

last_gen = df_random[df_random['Generation'] == 100]
sns.set(style="whitegrid")

fig, ax = plt.subplots(figsize=(17, 8))

sns.boxplot(x='Group', y='Fitness', data=last_gen, palette="Set3", ax=ax)
plt.title('Boxplot of Fitness Values for Each Group', fontsize=16)
plt.suptitle('')
plt.xticks(rotation=45, ha='right')
plt.show()
```
